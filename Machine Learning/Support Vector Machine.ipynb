**Train Using Support Vector Machine (SVM)**

```python
from sklearn.model_selection import train_test_split
```

```python
df.head()
```

```python
X = df.drop(['target','flower_name','tartget'], axis='columns')
y = df.target
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
params = model.get_params()
print(params)
```

```python
model.score(X_test, y_test)
```

The parameters can be used to tune the model

```python
model_1 = SVC(C=10)
model_1.fit(X_train, y_train)
params_1 = model_1.get_params()
print(params_1)
```

```python
model_1.score(X_test, y_test)
```

The C of 10 made the model better

```python
model_2 = SVC(gamma=100) 
model_2.fit(X_train, y_train)
params_2 = model_1.get_params()
print(params_2)
```

```python
model_2.score(X_test, y_test)
```

A gamma of 10 makes the model worse
