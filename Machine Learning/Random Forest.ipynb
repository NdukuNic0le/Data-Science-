```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.gray() 
for i in range(4):
    plt.matshow(digits.images[i]) 
```

```python
digits.data[:5]
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
digits.target
```

```python
 len(df)
```

```python
#Let's add a target column to our df
df['target'] = digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'], axis = 1), digits.target, test_size = 0.2)
```

```python
len(X_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
params = model.get_params()
print(params)
```

```python
model.score(X_test, y_test)
```

When 'n_estimators' = 20, model score = 98.0555%
<br>
When 'n_estimators' = 100, model score = 98.6111%
<br>
The number of estimators are the number of trees, more trees, higher score

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print (classification_report(y_test, y_predicted))
```

```python
cm = confusion_matrix (y_test, y_predicted)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize= (10,7))
sns.heatmap(cm, annot=True, cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
