**Dropout Regularization for Deep Learning Neural Network**
<br>
The task is to train a network to discriminate between sonar signals bounced off a metal cylinder and those bounced off a roughly cylindrical rock
<br>
Dataset information: https://archive.ics.uci.edu/ml/datasets/Connectionist+Bench+(Sonar,+Mines+vs.+Rocks) Download it from here: https://archive.ics.uci.edu/ml/machine-learning-databases/undocumented/connectionist-bench/sonar/sonar.all-data
<br>
It is a binary classification problem that requires a model to differentiate rocks from metal cylinders. (column 60)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("sonar_dataset.csv", header = None)
df.head(5)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df[60].value_counts()
```

```python
X = df.drop(60, axis = 'columns')
y = df[60]
y.head()
```

Let's give column 60 numeric values , 0 or 1 using one hot encoding

```python
y = pd.get_dummies( y, drop_first=True )
y.sample(5) 
```

```python
# 1 is R, 0 is M
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_test.shape
```

**Deep Learning Model (ANN) without Dropout Layer**

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'), #input is 60 because we have 60 columns, 60 neurons in the hidden layer to start
    keras.layers.Dense(30, activation='relu'), # 30 neurons in the 2nd hidden layer, relu is the preferred activation for all hidden layers
    keras.layers.Dense(15, activation='relu'), # 15 neurons in the 3rd hidden layer
    keras.layers.Dense(1, activation='sigmoid') # 1 neuron in the output layer
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100, batch_size=8)
```

We intentionally overfitted to test it

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print (classification_report (y_test,y_pred))
```

**Deep Learning Model (ANN) with Dropout Layer**

```python
modeld = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid') 
])

modeld.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

modeld.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
modeld.evaluate(X_test, y_test)
```

```python
y_pred = modeld.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
```
